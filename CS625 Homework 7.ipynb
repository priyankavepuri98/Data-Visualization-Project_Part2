```python
data = pd.read_csv('hw7_question1.csv')
```

```python
#To find unique years
data.YEAR.unique()
```

```python
#Removing columns which are not required
df = data.drop(columns=['YEAR', 'POPULATION'])
```

```python
df.info()
```

```python
#Creating new df grouping SITE and SEX
siteGroup = df.groupby(['SITE', 'SEX'])
```

```python
siteGroup.describe()
```

```python
#Calculating sum of each combination pair of site and sex
siteSum = siteGroup.sum()
```

```python
#Creating a new dataframe for storing result
result = pd.DataFrame(columns=['SITE','SEX'])
```

```python
#Loop for appending all the values in a dataframe
i=0
for site in data.SITE.unique():
    for sex in data.SEX.unique():
        result = result.append({'SITE': site, 'SEX':sex, 'SUM':siteSum.COUNT.values[i]}, ignore_index=True)
        i=i+1
```

```python
result
```

```python
result.to_csv(path_or_buf='HW7.csv', index=False)
```
